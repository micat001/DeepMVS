```python
%load_ext autoreload
%autoreload 2

import os
import numpy as np
import glob
from matplotlib import pyplot as plt
import cv2
import time
from matplotlib.colors import Normalize
import glob

normalizer = Normalize(vmin=1e-6, vmax=200)
depth_normalizer = Normalize(vmin=1e-6, vmax=10)
```

```python
dataset_dir = '/data/users/data/MVS_DATA/bike_stereo/'
input_files = glob.glob(os.path.join(dataset_dir, 'dense/0/images', "*.png"))
input_files.sort()
print(input_files)

disparity_files = glob.glob(os.path.join(dataset_dir,'output', "*output.npy"))
disparity_files.sort()
print(disparity_files)
```

```python
inputs = [cv2.cvtColor(cv2.imread(f), cv2.COLOR_BGR2RGB) for f in input_files]

plt.rcParams['figure.figsize'] = [20, 15]
plt.title('Disparity')
plt.subplot(1, 2, 1)
plt.title('Input - Image 1', fontsize=16)
pc = plt.imshow(inputs[0])
plt.axis(False)

plt.subplot(1, 2, 2)
plt.title('Input - Image 2', fontsize=16)
pc = plt.imshow(inputs[1])
plt.axis(False)

# plt.savefig(f"disparity_results/hawaii_{idx}.jpg",
#                bbox_inches='tight', pad_inches=0.1)
plt.tight_layout()
plt.show()
```
